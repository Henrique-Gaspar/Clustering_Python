```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
from sklearn.metrics import classification_report
from sklearn.preprocessing import StandardScaler
import numpy as np
from scipy import stats
```

```python
dados = pd.read_csv('CC GENERAL.csv')
dados.head()
```

```python
dados = dados.drop("CUST_ID", 1)
```

```python
duplicados = dados[dados.duplicated(keep='first')]
print(duplicados)
```

```python
dados["CREDIT_LIMIT"].fillna(dados["CREDIT_LIMIT"].mean(), inplace =True)
```

```python
dados["MINIMUM_PAYMENTS"].fillna(dados["MINIMUM_PAYMENTS"].mean(), inplace=True)
```

```python
ax, fig = plt.subplots(figsize = (12,6))
sns.heatmap(dados.corr(), annot=True)
```

```python
dados.head()
```

```python
scaler = StandardScaler()
dados_scaled = scaler.fit_transform(dados)
```

```python
score = []
range_values = range(1,20)
for i in range_values:
    kmeans = KMeans(n_clusters = i)
    kmeans.fit(dados_scaled)
    score.append(kmeans.inertia_)
    
plt.plot(score, '+-r')
```

```python
cluster = KMeans(n_clusters=6, random_state = 0).fit(dados_scaled)
```

```python
previsoes = cluster.labels_
centroides = cluster.cluster_centers_
```

```python
np.unique(previsoes)
```

```python
pca = PCA(n_components = 2)
dados_pca = pca.fit_transform(dados_scaled)
```

```python
dados_pca = pd.DataFrame(dados_pca, columns = ["PCA1", 'PCA2'])
```

```python
dados_pca = pd.concat([dados_pca, pd.DataFrame({"Grupo" : previsoes})], axis=1)
```

```python
ax, fig = plt.subplots(figsize=(18,12))
sns.scatterplot(x = dados_pca['PCA1'], y = dados_pca['PCA2'], hue = dados_pca['Grupo'], palette  = 'hls', alpha=0.5)
plt.title("Cluster")
plt.show()
```
